```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

#this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (15,10)
```

```python
# draw a 100k points from a Gaussian distribution
d = stats.norm.rvs(loc = 3.0, scale = .01, size = 100000)
```

```python
# sets up the plot as a 1x1, and gets the axes
fig, ax = plt.subplots(1, 1)
# plots a historgram of the data fromm above
ax.hist(d,50, density=True)
# sets the lables, guidelines, and ticks
plt.tick_params(labelsize = 24)
# sets the x limit of the plot (left, right)
plt.xlim([2.95,3.05])
# gets 1000 evenly spaced numbers between 2.95 and 3.05
x = np.linspace(2.95,3.05,1000)
# plots Gaussian curve
ax.plot(x,stats.norm.pdf(x,loc = 3, scale = 0.01),linewidth = 8,alpha = 0.7)
# shows plot
plt.show()
```

```python
# creates a histogram from the distribution data
ax = plt.hist(d,50, density=True)
# sets the plot to a semi-log plot in the y direction
plt.yscale('log')
# sets the size of the labels for the plot
plt.tick_params(labelsize = 24)
# sets the limit for the x axis
plt.xlim([2.95,3.05])
# shows the plot
plt.show()
```

```python
# sets up the plot as a 1x1, and gets the axes
fig, ax = plt.subplots(1, 1)
# graphs the histogram on the plot
ax.hist(d,50, density=True)
# makes the y axis log scaled
plt.yscale('log')
# sets the label size
plt.tick_params(labelsize = 24)
# sets the x axis range
plt.xlim([2.95,3.05])
# creates evenly spaced numbers
x = np.linspace(2.95,3.05,1000)
# adds a LOBF to the plot
ax.plot(x,stats.norm.pdf(x,loc = 3., scale = 0.01),linewidth = 8,alpha = 0.7)
# shows it all
plt.show()
```
