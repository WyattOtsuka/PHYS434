# Problem 0

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,np.pi*100,100000)
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=100000)*np.sqrt(N)+N
```

# Problem 1

## 1a) Make a scatter plot

```python
plt.scatter(x = t[:1000], y = bkgd[:1000], alpha = 0.5)
plt.xlabel("Time")
plt.ylabel("Background Data Values")
plt.title("Backgrd Data vs Time", fontsize = 24)
plt.show()
```

## 1b) Generalize your code

```python
def plotScatter(x, y, title, xlabel, ylabel, start = None, stop = None, total = None, alpha = 0.5, itr = 1):
    if stop == None and total != None:
        stop = start + total
    if start == None and total != None: 
        start = stop - total
    if start == None and total == None:
        start = 0
    if stop == None and total == None:
        stop = len(x)
    
    plt.scatter(x = x[start:stop:itr], y = y[start:stop:itr], alpha = 0.5)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title( title, fontsize = 24)
    plt.show()
```

```python
plotScatter(t,bkgd, "2000-5000 of Background Data vs Time", "Time", "Background Data Values", start = 2000, stop = 5000)
plotScatter(t,bkgd, "30 Entries Starting at 2000 of Background Data vs Time", "Time", "Background Data Values", start = 2000, total = 30)
plotScatter(t,bkgd, "600 Entries Stopping at 5000 of Background Data vs Time", "Time", "Background Data Values", stop = 5000, total = 600)
```

The first graph shows the data from 2000 to 5000, while the second shows 30 entries, starting at 2000, and the final shows 600 entries, stopping at 5000.

## 1c) Plot every 100th point

```python
plotScatter(t,bkgd, "Every 100th point of Background Data vs Time", "Time", "Background Data Values", itr = 100)
```

Here we can see the iterator is set to 100, meaning that only every 100th point is plotted on the graph

# Problem 2

## 2a) Make a 2d histogram plot

```python
fig, ax = plt.subplots(1,1)
fig.colorbar(plt.cm.ScalarMappable(), ax=ax)
plt.hist2d(t,bkgd,bins = 30)
plt.xlabel("Time")
plt.ylabel("Background Data Density")
plt.title("Backgrd Data vs Time", fontsize = 24)
plt.show()
```

# 2b) Clearly explain what is being plotted

In the above plot, we are seeing the density of the points within certain ranges. We see that the most common point is around 10, and it seems that there is an even distribution both above and below 10. The closer to blue the color is, the lower the density in that region. The closer to yellow that the bin color is, the higher the density.
